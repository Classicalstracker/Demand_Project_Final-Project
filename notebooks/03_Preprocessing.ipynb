```python
import pandas as pd
import sys
import os

# Add project root to path
sys.path.append('d:/Final Project')

from src.io import load_product_information, load_weekly_sales, load_external_features
from src.preprocess import preprocess_product_info, preprocess_weekly_sales, preprocess_external_features
from src.config import DATA_PROCESSED

print("Imports successful")
```

## Load Raw Datasets

```python
print("Loading Product_Information...")
df_product_raw = load_product_information()
print(f"Shape: {df_product_raw.shape}")
print(f"Columns: {list(df_product_raw.columns)}")

print("\nLoading Weekly_Sales...")
df_sales_raw = load_weekly_sales()
print(f"Shape: {df_sales_raw.shape}")
print(f"Columns: {list(df_sales_raw.columns)}")

print("\nLoading External_Features...")
df_external_raw = load_external_features()
print(f"Shape: {df_external_raw.shape}")
print(f"Columns: {list(df_external_raw.columns)}")
```

## Preprocess Product Information

```python
print("Preprocessing Product_Information...")
print(f"Before: {df_product_raw.shape}")

df_product_clean = preprocess_product_info(df_product_raw)

print(f"After: {df_product_clean.shape}")
print(f"Duplicates removed: {df_product_raw.shape[0] - df_product_clean.shape[0]}")

# Missing values summary
print("\nMissing values:")
print(df_product_clean.isnull().sum())

print("\nFirst few rows:")
df_product_clean.head()
```

### Insights: Product_Information
- Dataset shape changed from X to Y rows
- Z duplicate rows were removed
- Missing values in columns: [list if any]
- Data types validated and converted appropriately

## Preprocess Weekly Sales

```python
print("Preprocessing Weekly_Sales...")
print(f"Before: {df_sales_raw.shape}")

df_sales_clean = preprocess_weekly_sales(df_sales_raw)

print(f"After: {df_sales_clean.shape}")
print(f"Duplicates removed: {df_sales_raw.shape[0] - df_sales_clean.shape[0]}")

# Check uniqueness
unique_combinations = df_sales_clean.groupby(['Product_ID', 'Week_End_Date']).size()
duplicate_combos = (unique_combinations > 1).sum()
print(f"Duplicate Product_ID + Week_End_Date combinations: {duplicate_combos}")

# Missing values summary
print("\nMissing values:")
print(df_sales_clean.isnull().sum())

print("\nFirst few rows:")
df_sales_clean.head()
```

### Insights: Weekly_Sales
- Dataset shape changed from X to Y rows
- Z duplicate rows were removed
- Ensured uniqueness of Product_ID + Week_End_Date combinations
- Data sorted by Product_ID and Week_End_Date
- Missing values in columns: [list if any]

## Preprocess External Features

```python
print("Preprocessing External_Features...")
print(f"Before: {df_external_raw.shape}")

df_external_clean = preprocess_external_features(df_external_raw)

print(f"After: {df_external_clean.shape}")
print(f"Duplicates removed: {df_external_raw.shape[0] - df_external_clean.shape[0]}")

# Missing values summary
print("\nMissing values:")
print(df_external_clean.isnull().sum())

print("\nFirst few rows:")
df_external_clean.head()
```

### Insights: External_Features
- Dataset shape changed from X to Y rows
- Z duplicate rows were removed
- Season text standardized to title case
- Data sorted by Week_End_Date
- Missing values in columns: [list if any]

## Save Cleaned Datasets

```python
# Save to CSV files
output_dir = DATA_PROCESSED
output_dir.mkdir(parents=True, exist_ok=True)

df_product_clean.to_csv(output_dir / 'product_info_clean.csv', index=False)
df_sales_clean.to_csv(output_dir / 'weekly_sales_clean.csv', index=False)
df_external_clean.to_csv(output_dir / 'external_features_clean.csv', index=False)

print("Cleaned datasets saved successfully:")
print("- product_info_clean.csv")
print("- weekly_sales_clean.csv") 
print("- external_features_clean.csv")
```
