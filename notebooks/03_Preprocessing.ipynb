# Preprocessing

This notebook performs data cleaning and preprocessing on all three datasets.

## Import Required Libraries

```python
import pandas as pd
from src.io import load_product_information, load_weekly_sales, load_external_features
from src.preprocess import preprocess_product_information, preprocess_weekly_sales, preprocess_external_features
from src.config import DATA_PROCESSED
```

## Load Raw Datasets

```python
product_info_raw = load_product_information()
weekly_sales_raw = load_weekly_sales()
external_features_raw = load_external_features()

print("Raw datasets loaded.")
print(f"Product Information shape: {product_info_raw.shape}")
print(f"Weekly Sales shape: {weekly_sales_raw.shape}")
print(f"External Features shape: {external_features_raw.shape}")
```

## Preprocess Product Information

```python
print("Processing Product Information...")
product_info_clean = preprocess_product_information(product_info_raw)
print(f"Cleaned shape: {product_info_clean.shape}")
print(f"Cleaned columns: {list(product_info_clean.columns)}")
print("\nData types:")
print(product_info_clean.dtypes)
print("\nFirst few rows:")
display(product_info_clean.head())
```

### Insights: Product Information Preprocessing
- Column names have been trimmed to remove whitespace.
- Launch_Date has been converted to datetime format.
- Duplicates have been removed.
- Text columns have been standardized to lowercase.
- MSRP and Wholesale_Price are validated as numeric values.

## Preprocess Weekly Sales

```python
print("Processing Weekly Sales...")
weekly_sales_clean = preprocess_weekly_sales(weekly_sales_raw)
print(f"Cleaned shape: {weekly_sales_clean.shape}")
print(f"Cleaned columns: {list(weekly_sales_clean.columns)}")
print("\nData types:")
print(weekly_sales_clean.dtypes)
print("\nFirst few rows:")
display(weekly_sales_clean.head())
```

### Insights: Weekly Sales Preprocessing
- Week_End_Date has been converted to datetime format.
- Duplicates have been removed.
- Product_ID + Week_End_Date uniqueness has been validated.
- Qty_Sold, Inventory, On_Order, Returns, and Discount have been converted to numeric.
- Data is sorted by Product_ID and Week_End_Date for easier analysis.

## Preprocess External Features

```python
print("Processing External Features...")
external_features_clean = preprocess_external_features(external_features_raw)
print(f"Cleaned shape: {external_features_clean.shape}")
print(f"Cleaned columns: {list(external_features_clean.columns)}")
print("\nData types:")
print(external_features_clean.dtypes)
print("\nFirst few rows:")
display(external_features_clean.head())
```

### Insights: External Features Preprocessing
- Week_End_Date has been converted to datetime format.
- Rainfall and Avg_Temperature have been converted to numeric values.
- Holiday and Promo_Event have been validated as binary indicators.
- Season text has been standardized to lowercase.
- Duplicates have been removed.

## Save Cleaned Datasets

```python
# Ensure output directory exists
DATA_PROCESSED.mkdir(parents=True, exist_ok=True)

# Save cleaned datasets
product_info_path = DATA_PROCESSED / 'product_info_clean.csv'
weekly_sales_path = DATA_PROCESSED / 'weekly_sales_clean.csv'
external_features_path = DATA_PROCESSED / 'external_features_clean.csv'

product_info_clean.to_csv(product_info_path, index=False)
weekly_sales_clean.to_csv(weekly_sales_path, index=False)
external_features_clean.to_csv(external_features_path, index=False)

print(f"Saved: {product_info_path}")
print(f"Saved: {weekly_sales_path}")
print(f"Saved: {external_features_path}")
print("\nPreprocessing complete!")
```

## Preprocessing Summary

```python
print("\n=== PREPROCESSING SUMMARY ===")
print(f"\nProduct Information:")
print(f"  Raw rows: {product_info_raw.shape[0]}, Cleaned rows: {product_info_clean.shape[0]}")
print(f"  Columns: {product_info_clean.shape[1]}")
print(f"\nWeekly Sales:")
print(f"  Raw rows: {weekly_sales_raw.shape[0]}, Cleaned rows: {weekly_sales_clean.shape[0]}")
print(f"  Columns: {weekly_sales_clean.shape[1]}")
print(f"\nExternal Features:")
print(f"  Raw rows: {external_features_raw.shape[0]}, Cleaned rows: {external_features_clean.shape[0]}")
print(f"  Columns: {external_features_clean.shape[1]}")
print(f"\nAll cleaned datasets have been saved to data/processed/")
```
