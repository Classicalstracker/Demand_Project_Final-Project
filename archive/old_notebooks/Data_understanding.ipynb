```python
product_info.head()
product_info['Category'].unique()
product_info['Sub_Category'].unique()
product_info['Gender'].unique()
```

```python
product_info_clean = product_info.copy()

# Remove full duplicate rows (safe step)
product_info_clean = product_info_clean.drop_duplicates()

# Ensure Product_ID is unique
assert product_info_clean['Product_ID'].duplicated().sum() == 0, "Duplicate Product_ID found!"

# Convert Launch_Date if exists
if 'Launch_Date' in product_info_clean.columns:
    product_info_clean['Launch_Date'] = pd.to_datetime(product_info_clean['Launch_Date'], errors='coerce')

print("Product Info is clean and ready ✅")
```

## WEEKLY_SALES PREPROCESSING

```python
weekly_sales_clean = weekly_sales.copy()
weekly_sales_clean['Week_End_Date'] = pd.to_datetime(weekly_sales_clean['Week_End_Date'], errors='coerce')
weekly_sales_clean.isnull().sum()
weekly_sales_clean.duplicated().sum()
weekly_sales_clean = weekly_sales_clean.drop_duplicates()
weekly_sales_clean.duplicated(subset=['Product_ID', 'Week_End_Date']).sum()
weekly_sales_clean = weekly_sales_clean.drop_duplicates(subset=['Product_ID', 'Week_End_Date'])
weekly_sales_clean = weekly_sales_clean.sort_values(['Product_ID', 'Week_End_Date']).reset_index(drop=True)
weekly_sales_clean.describe()
```

```python
weekly_sales_clean.columns
numeric_cols = weekly_sales_clean.select_dtypes(include=[np.number]).columns
numeric_cols
```

```python
for col in numeric_cols:
    negative_count = (weekly_sales_clean[col] < 0).sum()
    print(f"{col}: {negative_count} negative values")
```

```python
weekly_sales_clean['Week_End_Date'].isnull().sum()
```

```python
weekly_sales_clean.groupby('Product_ID')['Week_End_Date'].nunique().describe()
```

```python
weekly_sales_clean.groupby('Product_ID')['Week_End_Date'].nunique().value_counts()
```

```python
print("========== WEEKLY SALES CLEANING SUMMARY ==========")
print("Original Shape:", weekly_sales.shape)
print("Cleaned Shape:", weekly_sales_clean.shape)
print("Duplicate Product-Week Rows:", weekly_sales_clean.duplicated(subset=['Product_ID', 'Week_End_Date']).sum())
print("Missing Values:\n", weekly_sales_clean.isnull().sum())
print("Date Range:", weekly_sales_clean['Week_End_Date'].min(), "to", weekly_sales_clean['Week_End_Date'].max())
```

```python
print("========== FINAL PREPROCESSING SUMMARY ==========")
print("Product Info Clean Shape:", product_info_clean.shape)
print("Weekly Sales Clean Shape:", weekly_sales_clean.shape)
print("Unique Products in Product Info:", product_info_clean['Product_ID'].nunique())
print("Unique Products in Weekly Sales:", weekly_sales_clean['Product_ID'].nunique())
print("Weekly Sales Date Range:", weekly_sales_clean['Week_End_Date'].min(), "to", weekly_sales_clean['Week_End_Date'].max())
```
